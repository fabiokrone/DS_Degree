```python
import numpy as np
import pandas as pd
from scipy import stats
import math
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
#Baixando os dataframes
wq_r = pd.read_csv('winequality-red.csv', sep= ';')
wq_w = pd.read_csv('winequality-white.csv', sep= ';')
```

```python
#Dataframe vinho tinto
wq_r.head()
```

```python
#Criando uma coluna para definir o tipo do vinho
wq_r2=wq_r.assign(types='red')
wq_r2
```

```python
#Dataframe vinho branco
wq_w.head()
```

```python
# Criando uma coluna para definir o tipo do vinho
wq_w2=wq_w.assign(types='white')
wq_w2
```

```python
#Unindo os 2 dataframes
wq_rw1 = pd.merge(wq_w2, wq_r2, how='outer')
```

```python
#Dataframes associados
wq_rw1
```

```python
# Ordenando pelo nível da Qualidade
wq_rw = wq_rw1.sort_values(by='quality')
wq_rw
```

```python
# Usando o describe 
wq_rw.describe()
```

```python
# Verificando se há dados nulos e quais tipos de dados
wq_rw.info()
```

```python
# Testando se não foi perdida informação na união dos dataframes
wq_sum_rows = wq_r.shape[0] + wq_w.shape[0]
print(f'As 2 tabelas tem juntas o total de {wq_sum_rows} linhas')
```

```python
# Verificando todos os números classificatórios de qualidade
wq_rw['quality'].unique()
```

```python
# Filtrando todas as linhas cuja qualidade seja igual ou superior a 6
wq_rw.loc[wq_rw['quality'] >= 6]
```

```python
# Filtrando quantos vinhos vermelhos estão com qualidade igual ou superior a 6 
wq_rw.loc[(wq_rw['quality'] >= 6) & (wq_rw['types'] == 'red')]
```

```python
# Filtrando quantos vinhos brancos estão com qualidade igual ou superior a 6 
wq_rw.loc[(wq_rw['quality'] >= 6) & (wq_rw['types'] == 'white')]
```

```python
# Tirando a média das características dos vinhos pelo nível de qualidade
wq_rw_q = wq_rw.groupby("quality").mean()
wq_rw_q
```

```python
# OBSERVANDO É POSSÍVEL PERCEBER VARIAÇÕES COM CERTO PADRÃO NAS COLUNAS: VOLATILE ACIDITY / CITRIC ACID / CHLORIDES
```

```python
# Comparando qualidade com a média de volatile acidity. 
wq_rw_q.plot(y='volatile acidity',kind='bar',figsize=(6,6))
```

#### É possível perceber no gráfico acima que na média os vinhos brancos e o tinto tem qualidade igual ou maior que 6 quando a o volatile acidity está igual ou abaixo de 0.3

```python
# Comparando qualidade com a média de citric acid. 
wq_rw_q.plot(y='citric acid',kind='bar',figsize=(6,6))
```

#### No gráfico acima mostra que na média os vinhos que são de nivel 6 para cima tem um nível de ácido cítrico acima de 0.30.


```python
# Comparando qualidade com a média de chlorides. 
wq_rw_q.plot(y='chlorides',kind='bar',figsize=(6,6))
```

#### No gráfico acima mostra que na média os vinhos que são de nivel 6 para cima tem um nível chlorides abaixo de 0.056. 

## OBSERVAÇÃO:
  Necessário se atentar ao fato de que quando tiramos a média perdemos vários valores no caminho. Principalmente os outliers, podemos usar esses gráficos para fazer uma análise mais apurada dos índices mais perceptíveis de padrão. No caso os três mensionados acima.
  
  ##### TAMBEM TEMOS A QUESTÃO DOS PRODUTOS ENTRE A QUALIDADE NÍVEL 7 E 8. ISTO PARA volatile acidity	E PARA citric acid. Esses casos quebram o gráfico, a lógica não se aplica perfeitamente. Precisamos verificar qual a alternativa para sanar essas questões, talvez aplicando técnicas voltadas para cada caso isolado seja visualmente melhor. Isto sem usar média, trabalhando com outras métricas.

#### Mais visualização

Separando os valores de qualidade em três grupos diferentes, para que possamos fazer as coisas um pouco mais fáceis:
1. <b>Low:</b> contendo vinhos com qualidade de 3 ou 4
2. <b>Medium:</b> contendo vinhos com qualidade de 5,6 ou 7
3. <b>High:</b> contendo vinhos com qualidade de 8 ou 9

```python
#Fizemos uma cópia do dataframe e qualidade de grupo em diferentes grupos
df_aux = wq_rw1.copy()
df_aux['quality'].replace([3,4],['low','low'],inplace=True)
df_aux['quality'].replace([5,6,7],['med','med','med'],inplace=True)
df_aux['quality'].replace([8,9],['high','high'],inplace=True)

sns.countplot(df_aux['quality'])
```

```python
#Colocamos alguns histogramas que mostram os valores das características selecionadas
flistt = ['alcohol','sulphates','citric acid','volatile acidity']
low = df_aux[df_aux['quality'] == 'low']
medium = df_aux[df_aux['quality'] == 'med']
high = df_aux[df_aux['quality'] == 'high']
plt.rcParams.update({'font.size': 8})
plot, graphs = plt.subplots(nrows= 2, ncols= 2, figsize=(12,6))
graphs = graphs.flatten()
for i, graph in enumerate(graphs):
    graph.figure
    binwidth= (max(df_aux[flistt[i]]) - min(df_aux[flistt[i]]))/30
    bins = np.arange(min(wq_rw1[flistt[i]]), max(df_aux[flistt[i]]) + binwidth, binwidth)
    graph.hist([low[flistt[i]],medium[flistt[i]],high[flistt[i]]], bins=bins, alpha=0.6, density=True, label=['Low','Medium','High'], color=['red','green','blue'])
    graph.legend(loc='upper right')
    graph.set_title(flistt[i])
plt.tight_layout()
```

Como podemos ver nos histogramas, valores mais elevados de álcool, sulfatos e ácido cítrico parecem pertencer a vinhos de maior qualidade enquanto valores mais elevados de cidade volátil estão presentes em vinhos de qualidade inferior.
